```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('..\datasets\heart.csv')
data.drop("Unnamed: 0",axis=1,inplace=True)
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Ca'] = data['Ca'].fillna(data['Ca'].median())
data['Thal'] = data['Thal'].fillna(data['Thal'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
# import seaborn as sns

# corr = data.corr()

# plt.figure(figsize = (15,15))
# sns.heatmap(corr, annot = True)
```

```python
data['AHD'] = data['AHD'].map({'Yes': 1, 'No': 0})
data['AHD']=data['AHD'].astype('int')
data.info()
```

```python
categorical_columns = data.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_columns
```

```python
data['ChestPain'].value_counts()
```

```python
data['Thal'].value_counts()
```

```python
data['AHD'].value_counts()
```

```python
df = pd.get_dummies(data, columns=categorical_columns)

for column in df.select_dtypes(include=['bool']).columns:
    df[column] = df[column].astype(int)

df
```

```python
X = df.drop(['AHD'], axis = 1)
y = df['AHD']
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X_train_scaled)
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
cumulative_variance_ratio
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(criterion='gini', n_estimators=20)
model.fit(X_train_pca, y_train)
y_pred = model.predict(X_test_pca)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred=y_pred, y_true=y_test)*100)
```

```python
import pickle
pickle.dump(model, open("../models/heart-model.pkl",'wb'))
pickle.dump(scaler, open("../scalar/heart-scaler.pkl",'wb'))
pickle.dump(pca, open("../pca/heart-pca.pkl",'wb'))
```
