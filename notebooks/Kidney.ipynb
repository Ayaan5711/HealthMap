```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('..\datasets\kidney_disease.csv')
data.head()
```

```python
data.classification=data.classification.replace("ckd\t","ckd") 
```

```python
data['classification'] = data['classification'].replace(['ckd','notckd'], [1,0])
```

```python
df = data.dropna(axis = 0)
print(f"Before dropping all NaN values: {data.shape}")
print(f"After dropping all NaN values: {df.shape}")
```

```python
df.index = range(0,len(df),1)
df
```

```python
for i in df['wc']:
    print(i)
```

```python
df['wc']=df['wc'].replace(["\t6200","\t8400"],[6200,8400])
```

```python
for i in df['wc']:
    print(i)
```

```python
df.info()
```

```python
df['pcv']=df['pcv'].astype(int)
df['wc']=df['wc'].astype(int)
df['rc']=df['rc'].astype(float)
df.info()
```

```python
object_dtypes = df.select_dtypes(include = 'object')
object_dtypes.head()
```

```python
dictonary = {
        "rbc": {
        "abnormal":1,
        "normal": 0,
    },
        "pc":{
        "abnormal":1,
        "normal": 0,
    },
        "pcc":{
        "present":1,
        "notpresent":0,
    },
        "ba":{
        "notpresent":0,
        "present": 1,
    },
        "htn":{
        "yes":1,
        "no": 0,
    },
        "dm":{
        "yes":1,
        "no":0,
    },
        "cad":{
        "yes":1,
        "no": 0,
    },
        "appet":{
        "good":1,
        "poor": 0,
    },
        "pe":{
        "yes":1,
        "no":0,
    },
        "ane":{
        "yes":1,
        "no":0,
    }
}
```

```python
df=df.replace(dictonary)
df.head()
```

```python
df['classification'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE, RandomOverSampler

# Initialize SMOTE
smote = SMOTE(random_state=1)

# Apply SMOTE to the training data
X_resampled_smote, y_resampled_smote = smote.fit_resample(df.drop('classification',axis=1), df['classification'])

# Initialize RandomOverSampler
ros = RandomOverSampler(random_state=1)

# Apply Random Over-Sampling to the SMOTE-resampled data
X_resampled_final, y_resampled_final = ros.fit_resample(X_resampled_smote, y_resampled_smote)

X, y = X_resampled_final, y_resampled_final
```

```python
X.columns.tolist()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X_train_scaled)
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
cumulative_variance_ratio
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=20)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(criterion='gini', n_estimators=20)
model.fit(X_train_pca, y_train)
y_pred = model.predict(X_test_pca)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred=y_pred, y_true=y_test)*100)
```

```python
import pickle
pickle.dump(model, open("../models/kidney-model.pkl",'wb'))
pickle.dump(scaler, open("../scalar/kidney-scaler.pkl",'wb'))
pickle.dump(pca, open("../pca/kidney-pca.pkl",'wb'))
```
